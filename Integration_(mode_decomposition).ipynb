We are considering the following functions:
$$
f_1(x)=e^x\cos(x)
$$
in $I=[0,\pi/2]$,

$$
f_2(x)=e^x
$$
in $I=[-1,3]$,

$$
f_3(x)=
\begin{cases} 
  e^{2x}   &\text{if } x < 0       \\ 
  x-2\cos(x)+4   &\text{if } x\geq0 \\ 
\end{cases} 
$$
in $I=[-1,1]$.

We applied Simpson's 3/8 rule and Gauss-Legendre quadrature. 
Along with Simpson's 3/8 rule we also applied Simpson's 1/3 rule and the trapezoidal rule to deal with the cases MOD(m, 3)=1, and MOD(m, 3)=2, where m is the number of subintervals.

```python
import numpy as np
import matplotlib.pyplot as plt

# Loop through the three files
for item in range(1,4):
    filename = f"doc_{item}.txt"
    data = np.loadtxt(filename)
    x, y = data[:, 0], data[:, 1]

    plt.figure()
    plt.plot(x, y)
    plt.xlabel("log(h)")
    plt.ylabel("Error")
    plt.title(f"Function {item}")
    #plt.legend()
    plt.grid()
    plt.show()

for item in range(4,7):
    filename = f"doc_{item}.txt"
    data = np.loadtxt(filename)
    x, y = data[:, 0], data[:, 1]

    plt.figure()
    plt.plot(x, y)
    plt.xlabel("log(h)")
    plt.ylabel("Error")
    plt.title(f"Function {item-3}")
    #plt.legend()
    plt.grid()
    plt.show()
```
